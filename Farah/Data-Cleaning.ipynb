```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import train and test data:
train = pd.read_csv('/Users/farahayyad/Desktop/SLR-Research/SLR-Data/Titanic-Data/train.csv')
test = pd.read_csv('/Users/farahayyad/Desktop/SLR-Research/SLR-Data/Titanic-Data/test.csv')
```

# Removing Missing Values

```python
# keep rows with at least 11 non-NAN values (remove rows with (13-11) 2 missing values)
clean_D = train.dropna(thresh=11)

# Clean up the table then display it: 
clean_D.reset_index(inplace=True)
clean_D = clean_D.drop(labels=["index"], axis=1)
clean_D
```

# Removing Unused Columns:

Since only: ['Age', 'Survived', 'Sex', 'Pclass', 'Parch', 'Embarked', 'SibSp'] will be used, i will drop everything else

```python
print(clean_D.keys())
dataset = clean_D.drop(columns=['Cabin','Name','PassengerId','Fare','Ticket'])
dataset
```

# Encoding: 

# 1. Age:

```python
dataset.loc[dataset.Age < 15,'Age'] = 0
dataset.loc[(dataset.Age >= 15) & (dataset.Age < 55), 'Age'] = 1
dataset.loc[dataset.Age >= 55, 'Age'] = 2
dataset
```

# 2. Sex:

```python
dataset.loc[dataset.Sex == 'male', 'Sex'] = 0
dataset.loc[dataset.Sex == 'female', 'Sex'] = 1
dataset
```

# 3. Embarked:

```python
dataset.loc[dataset.Embarked == 'C', 'Embarked'] = 0
dataset.loc[dataset.Embarked == 'Q', 'Embarked'] = 1
dataset.loc[dataset.Embarked == 'S', 'Embarked'] = 2
dataset
```

# 4. PClass:

```python
dataset['Pclass'] = dataset['Pclass'].astype(int)
dataset['Pclass'] = dataset['Pclass'] - 1
dataset
```

```python
dataset.to_csv('final_data.csv')
```
