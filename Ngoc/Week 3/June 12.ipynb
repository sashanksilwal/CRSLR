```python
train = pd.read_csv('titanic/train.csv')
test = pd.read_csv('titanic/test.csv')
gender_submission = pd.read_csv('titanic/gender_submission.csv')
```

```python
total_test = test.shape[0]
verify = []
for i in range(total_test):
    if gender_submission.iloc[i]['PassengerId'] == test.iloc[i]['PassengerId']:
        verify.append(True)
set(verify)
```

```python
test['Survived'] = gender_submission['Survived']
```

```python
# Inspect features in terms of unique values
def unique_values(data):
    print('Total number of passengers:', data.shape[0])
    for feature in data.columns:
        print("Number of unique values in {}: {}".format(feature, data[feature].unique().shape[0]))

def null_values(data):
    print("Number of null values:\n", data.isnull().sum())
```

```python
data = pd.concat([train, test])
```

```python
data
```

```python
unique_values(data)
null_values(data)
```

```python
data['Title']=data['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
data.Title.value_counts()
def girl(aa):
    if (not np.isnan(aa.Age))&(aa.Title=='Miss')&(aa.Age<=14):
        return 'Girl'
    elif (np.isnan(aa.Age))&(aa.Title=='Miss')&(aa.Parch!=0):
        return 'Girl'
    else:
        return aa.Title

data['Title']=data.apply(girl,axis=1)

Tit=['Mr','Miss','Mrs','Master','Girl','Rareman','Rarewoman']
for i in Tit:
    data.loc[(np.isnan(data.Age))&(data.Title==i),'Age']=data.loc[data.Title==i,'Age'].median()
# https://www.programmersought.com/article/74205101984/
```
