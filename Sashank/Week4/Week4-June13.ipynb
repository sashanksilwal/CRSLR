### Checking for missing data

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Fare', 'Embarked'])
```

```python
df.isnull().sum()
```

```python
df[df["Age"].isnull()]
```

```python

df['Title']=df['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
df.Title.value_counts()
def girl(aa):
    if (not np.isnan(aa.Age))&(aa.Title=='Miss')&(aa.Age<=14):
        return 'Girl'
    elif (np.isnan(aa.Age))&(aa.Title=='Miss')&(aa.Parch!=0):
        return 'Girl'
    else:
        return aa.Title

df['Title']=df.apply(girl,axis=1)

Tit=['Mr','Miss','Mrs','Master','Girl','Rareman','Rarewoman']
for i in Tit:
    df.loc[(np.isnan(df.Age))&(df.Title==i),'Age']=df.loc[df.Title==i,'Age'].median()
```

```python
df.isnull().sum()
```

#### Dropping 2 missing ages
